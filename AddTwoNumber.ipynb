```python
#Add two num.
import tkinter as tk
from tkinter import messagebox

class AdderApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Unlimited Adder")
        self.master.geometry("400x300") 

        # Create input fields
        self.label1 = tk.Label(master, text="Enter first number:")
        self.label1.pack()
        
        self.entry1 = tk.Entry(master)
        self.entry1.pack()

        self.label2 = tk.Label(master, text="Enter second number:")
        self.label2.pack()

        self.entry2 = tk.Entry(master)
        self.entry2.pack()

        # Create a button to add numbers
        self.add_button = tk.Button(master, text="Add", command=self.add_numbers)
        self.add_button.pack()

        # Create a label to display the all result
        self.result_label1 = tk.Label(master, text="Result_Two : 0")
        self.result_label1.pack()

        # Create a label to display the only two result
        self.result_label2 = tk.Label(master, text="Result_All: 0")
        self.result_label2.pack()

        # Initialize result variable
        self.result_Two = 0
        self.result_All =0

    def add_numbers(self):
        try:
            # Get input values and convert to float
            num1 = float(self.entry1.get())
            num2 = float(self.entry2.get())
            
            # Add numbers and update result
            self.result_All += num1 + num2
            self.result_Two = num1 + num2
            
            # Display result
            self.result_label1.config(text=f"Result_Two: {self.result_Two}")
            self.result_label2.config(text=f"Result_All: {self.result_All}")

            # Clear input fields
            self.entry1.delete(0, tk.END)
            self.entry2.delete(0, tk.END)

        except ValueError:
            messagebox.showerror("Input Error", "Please enter valid numbers.")

if __name__ == "__main__":
    root = tk.Tk()
    app = AdderApp(root)
    root.mainloop()
```
